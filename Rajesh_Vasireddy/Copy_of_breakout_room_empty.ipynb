### Question

Build a gradio UI where a user can supply a textual input and the output is the sentiment of that input.

### Instructions


1. figure out how to generate sentiment using a hf model - dont worry about gradio yet. you are free to use any hf model of your choice.
3. once summarization is figured, build the gradio UI using Interface!

```python
from transformers import pipeline

# Load the sentiment analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')

def get_sentiment(text):
  """Analyzes the sentiment of the given text."""
  result = sentiment_analyzer(text)[0]
  return f"Label: {result['label']}, Score: {result['score']:.4f}"

# Test the function
print(get_sentiment("I love using Hugging Face models!"))
print(get_sentiment("This is terrible."))
```

```python
import gradio as gr

# Create the Gradio interface
iface = gr.Interface(fn=get_sentiment, inputs="text", outputs="text",
                     title="Sentiment Analysis with Hugging Face",
                     description="Enter some text and get its sentiment.")

# Launch the interface
iface.launch()
```
